```python
# Import all the required Python Libraries.
import numpy as np
import pandas as pd
```

```python
# Load the Dataset into pandas' dataframe.
dataset = pd.read_csv('nba.csv')
```

```python
# Data Preprocessing

# 1. Check for missing values in the data 
print(dataset.isnull()) # This will return True if the value is missing and False if the value is not missing for each cell in the dataset.
print(dataset.isnull().sum()) # This will return the number of missing values in each column of the dataset.
```

```python
# 2. Get some initial statistics
dataset.describe() # This will return the initial statistics of the dataset.
```

```python
# 3. head() function returns the first n rows for the object based on position. It is useful for quickly testing if your object has the right type of data in it.
dataset.head() # This will return the first 5 rows of the dataset by default.
# dataset.head(10) # This will return the first 10 rows of the dataset.
```

```python
# 4. Provide variable descriptions. Types of variables etc. Check the dimensions of the dataframe.
print(dataset.dtypes) # This will return the data types(i.e., character, numeric, integer, factor, and logical) of each column in the dataset.
print(dataset.shape) # This will return the dimensions of the dataset.
```

```python
# Apply proper type conversions.
# If variables are not in the correct data type, apply proper type conversions.

# Convert the 'Height' column to float
dataset['Height'] = dataset['Height'].str.replace('-', '.')
dataset['Height'] = dataset['Height'].astype(float)

dataset.dtypes
```

```python
# Turn categorical variables into quantitative variables in Python.
# Categorical variables are variables that contain label values rather than numeric values.
# The number of possible values is often limited to a fixed set.

# In our nba dataset, the 'Position' column is a categorical variable(C, SG, SF, PF, PG). We make a new column for each category and assign a 1 or 0 to the column. 1 if the player is in that category and 0 if the player is not in that category
# This technique is called one hot encoding.

test = pd.get_dummies(dataset['Position'], dtype=int) # This will convert the 'Position' column into quantitative variables.
dataset = pd.concat([dataset, test], axis=1) # This will concatenate the 'Position' column with the dataset.
dataset.drop('Position', axis=1, inplace=True) # This will drop the 'Position' column from the dataset.

dataset.head() # This will return the first 5 rows of the dataset.
```
